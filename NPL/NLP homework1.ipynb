```python
 #3. Tokenization -  sozleri bolur
tokens = word_tokenize(text_cleaned)
print(f"3. Tokenization: {tokens}")

```

```python
# 4. Stopwords Removal -menasiz goruduymuz sozleri silme
stop_words = set(stopwords.words('english'))
filtered_tokens = [w for w in tokens if w not in stop_words]
print(f"4. Stopwords silindi: {filtered_tokens}")

```

```python
# 5. Lemmatization-sozleri kok halina salmaq
lemmatizer = WordNetLemmatizer()
lemmatized_output = [lemmatizer.lemmatize(w) for w in filtered_tokens]
print(f"5. Lemmatization: {lemmatized_output}")

```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
sentences = [
    "I love NLP",
    "NLP is amazing",
    "Machine learning is useful",
    "Deep learning is powerful"
]
```

```python
def full_preprocessing(text): ## birinci preprocessing hissesi
    # 1 Lowercase
    text = text.lower()
    ## 2 Punctuation remove
    text = text.translate(str.maketrans('', '', string.punctuation))
    # 3 Tokenization
    tokens = word_tokenize(text)
    # 4 Stopwords remove
    stop_words = set(stopwords.words('english'))
    tokens = [w for w in tokens if w not in stop_words]
    # 5Lemmatization
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(w) for w in tokens]

    return tokens

# Test edək
text = """
    I love NLP",
    "NLP is amazing",
    "Machine learning is useful",
    "Deep learning is powerful"""
print(full_preprocessing(text))
```

```python
##TF IDF hesablama

from sklearn.feature_extraction.text import TfidfVectorizer

import pandas as pd
processed_data = full_preprocessing(text)
# 1. Preprocessingden keciridyim dtaa
processed_data = [
    'love nlp',
    'nlp amazing',
    'machine learning useful',
    'deep learning powerful'
]


vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(processed_data)


feature_names = vectorizer.get_feature_names_out()

print("Words:",feature_names )
print("sozlerin cekisi):")
print(tfidf_matrix.toarray())
## en vacib sozler cekisi yuxari olanadir cox islendikce cekisi az olur burdan da en vacibleri love amazing useful deye bilerik
```

```python
#Word2Vec TF-IDF-den ferqli olaraq szoleri sadece saymır onların menasini ve bir biri ile elaqesini oyrenir hem de
import nltk
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import string
sentences = [
    "natural language processing is part of artificial intelligence",
    "machine learning helps computers learn from data",
    "deep learning is a subset of machine learning",
    "artificial intelligence is transforming technology",
    "neural networks are widely used in deep learning",
    "nlp techniques are used in chatbots and translation systems",
    "machine learning models learn patterns from data",
    "deep learning improves speech recognition systems"
]
```

```python
tokenized_sentences = [word_tokenize(s.lower()) for s in sentences] ##sade tokenize ledim
```

```python
##moodeli yaradaq
model = Word2Vec(sentences=tokenized_sentences, vector_size=100, window=5, min_count=1, workers=4)# window=5: etrafindaki 5 söze bax # vector_size=100: hər sözü 100 ededle ifadə et
```

```python
similar_words = model.wv.most_similar("machine", topn=5)

print("'machine' ssozuen yaxin olan sozler:")
for word, score in similar_words:
    print(f"{word}: {score:.4f}")
    ## en yaxnini datadir
```
