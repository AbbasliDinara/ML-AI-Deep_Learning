```python
#hamming
s1 = pd.Series([1, 0, 1, 1,0])
s2 = pd.Series([1, 1, 0, 1,1])

hamming = 0

if len(s1) != len(s2):
    print("Eyni uzunluqda olmalidir")
else:
    for i in range(len(s1)):
        if s1.iloc[i] != s2.iloc[i]:
            hamming += 1


hamming
```
