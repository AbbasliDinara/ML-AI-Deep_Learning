```python
from sklearn.mixture import GaussianMixture
gmm = GaussianMixture(n_components=1)
gmm.fit(data)
gmm.score_samples(data)
```

```python
outliers = np.quantile(gmm.score_samples(data),0.1)

outliers
data["isOutlier"] = gmm.score_samples(data) < outliers
```

```python
f1=np.mean(data["V1"])
f1_std=np.std(data["V1"])
f2=np.mean(data["V2"])
f2_std=np.std(data["V2"])
```

```python
def gaussion_prob(x,mean,std):
  exponent=np.exp(-((x-mean)**2)/(2*std**2))
  return(1/(np.sqrt(2*np.pi)*std))*exponent
```

```python
eps=0.000000000000001
```

```python
probabilities=gaussion_prob(data["V1"],f1,f1_std)*gaussion_prob(data["V2"],f2,f2_std)
```

```python
probabilities
```

```python
out=probabilities<eps
```

```python
out
```

```python

sns.scatterplot(data=data,x='V1', y='V2', hue=out, s=10)
```
